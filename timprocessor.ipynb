```python
import codecs, re
```

```python
f_corpus = codecs.open('datasets/MetsPreprocessedDone.txt', 'rU', encoding='utf-8')
corpus = f_corpus.readlines()
```

```python
corpus[:10]
```

```python
corp_list = []
corp_lens = []
for sent in corpus:
    corp_list.append(sent.split())
    corp_lens.append(len(sent.split()))
    
```

```python
corp_lens[:10]
```

```python
sum(corp_lens)/len(corp_lens)
```

```python
len([i for i in corp_lens if i < 13])
```

```python
shortcorp = [' '.join(i) for i in corp_list] # if len(i) < 13
```

```python
shortcorp
```

```python
import re
labels = []
newcorp = []
yes_met = False
for sent in shortcorp:
    sent = sent.strip()
    if re.search(r'(?<!NON)MET', sent):
        sent = re.sub(r'NONMET', '', sent)
        sent = re.sub(r'MET', '', sent)
        labels.append('yes')
    else:
        sent = re.sub(r'NONMET', '', sent)
        labels.append('no')
    newcorp.append(sent)
        
```

```python
for idx, sent in enumerate(newcorp[:10]):
    print(labels[idx], sent)
```

```python
print(len(newcorp))
print(len(labels))
print(len([i for i in labels if i == 'yes']))
```

```python
f_corpout = codecs.open('datasets/met_corpus.txt', 'w', encoding='utf-8')
f_labelout = codecs.open('datasets/met_labels.txt', 'w', encoding='utf-8')
for idx, sent in enumerate(newcorp):
    f_corpout.write(sent)
    f_corpout.write('\n')
    f_labelout.write(labels[idx])
    f_labelout.write('\n')
```
